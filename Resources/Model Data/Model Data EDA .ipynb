```python
mul_s_corr = mul_s.corr(method='spearman').abs()
mul_s_corr
```

```python
import numpy as np 

# Select upper triangle of correlation matrix
upper = mul_s_corr.where(np.triu(np.ones(mul_s_corr.shape), k=103).astype(np.bool))

# Find features with correlation greater than 0.95
to_drop = [column for column in upper.columns if any(upper[column] > 0.75)]

# Drop features 
mul_s_corr.drop(to_drop, axis=1, inplace=True)
```

```python
import pandas as pd 
from datetime import time
import matplotlib.pyplot as plt
import datetime as dt
from datetime import date
from datetime import timedelta
```

```python
# df = pd.read_csv('covid_deaths_us.csv', header=2)
df = pd.read_csv('updated_covid_deaths_us.csv', header=2)
df.head()
```

```python
df.info()
```

```python
df = df.drop(columns={'State'})
df
```

```python
df['Date New'] = pd.to_datetime(df['Date'])
df['Date New2'] = df['Date New'].apply(lambda x: x.strftime('%m/%d/%Y'))
df ['Date New2'] = pd.to_datetime(df['Date New'])

df.info()
```

```python
ndf = df.drop(columns={'Date', 'Date New'}, axis=1)
ndf = ndf.rename(columns={'Date New2':'Date'})
ndf
```

```python
print(ndf['Date'].min())
print(ndf['Date'].max())
df['Date'].count()
```

```python
# binary binning  of new deaths 

ndf ['Binary_Class'] = pd.cut(ndf['New Deaths'], [-1, 1200, 5000], labels=['Low', 'High'])
ndf.head()
```

```python
ndf['Binary_Class'].value_counts()
```

```python
ndf['Binary_Class'].count()
```

```python
binary_outcome_df = ndf.drop(columns={'New Deaths', '7-Day Moving Avg', 'Historic Deaths'})
binary_outcome_df
```

```python
ndf['New Deaths'].describe()
```

```python
ndf ['Multi_Class'] = pd.cut(ndf['New Deaths'], [-1, 549, 949, 5000], labels=['0-549', '550-1698', '1699-5000'])
ndf.head()
```

```python
ndf['Multi_Class'].value_counts()
```

```python
ndf['Multi_Class'].count()
```

```python
# (date + timedelta(days=4))
ndf['Search Date'] = ndf['Date'] - timedelta(days=10)
ndf.head()
```
